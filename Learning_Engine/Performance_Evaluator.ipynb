## _Performance Evaluation_

```python
sns.set_style("darkgrid")
sns.set_palette("viridis")
```

### _Relation between model's energy consumption and file size of image_

```python
fig, axs = plt.subplots(1, 4, figsize=(15, 4))

sns.scatterplot(data=nano_model, x="file_size(kB)", y="energy_consumption(J)", ax=axs[0])
axs[0].set_title("Nano Model")
axs[0].set_xlabel("File Size (kB)")
axs[0].set_ylabel("Energy Consumption (J)")

axs[0].set_ylim([0, 40])


sns.scatterplot(data=small_model, x="file_size(kB)", y="energy_consumption(J)", ax=axs[1])
axs[1].set_title("Small Model")
axs[1].set_xlabel("File Size (kB)")
axs[1].set_ylabel("Energy Consumption (J)")

axs[1].set_ylim([0, 40])

sns.scatterplot(data=med_model, x="file_size(kB)", y="energy_consumption(J)", ax=axs[2])
axs[2].set_title("Medium Model")
axs[2].set_xlabel("File Size (kB)")
axs[2].set_ylabel("Energy Consumption (J)")

axs[2].set_ylim([0, 40])

sns.scatterplot(data=large_model, x="file_size(kB)", y="energy_consumption(J)", ax=axs[3])
axs[3].set_title("Large Model")
axs[3].set_xlabel("File Size (kB)")
axs[3].set_ylabel("Energy Consumption (J)")

axs[3].set_ylim([0, 40])

plt.suptitle("Energy Consumption vs File Size (Individual Models)", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
sns.set_palette("Set2")
```

### _Relation between model's energy consumption and detection duration_

```python
fig, axs = plt.subplots(1, 4, figsize=(15, 4))

sns.scatterplot(data=nano_model, x="measure_duration(s)", y="energy_consumption(J)", ax=axs[0])
axs[0].set_title("Nano Model")
axs[0].set_xlabel("Measure Duration (s)")
axs[0].set_ylabel("Energy Consumption (J)")
axs[0].set_ylim([0, 40])
axs[0].set_xlim([0, 0.9])

sns.scatterplot(data=small_model, x="measure_duration(s)", y="energy_consumption(J)", ax=axs[1])
axs[1].set_title("Small Model")
axs[1].set_xlabel("Measure Duration (s)")
axs[1].set_ylabel("Energy Consumption (J)")
axs[1].set_ylim([0, 40])
axs[1].set_xlim([0, 0.9])

sns.scatterplot(data=med_model, x="measure_duration(s)", y="energy_consumption(J)", ax=axs[2])
axs[2].set_title("Medium Model")
axs[2].set_xlabel("Measure Duration (s)")
axs[2].set_ylabel("Energy Consumption (J)")
axs[2].set_ylim([0, 40])
axs[2].set_xlim([0, 0.9])

sns.scatterplot(data=large_model, x="measure_duration(s)", y="energy_consumption(J)", ax=axs[3])
axs[3].set_title("Large Model")
axs[3].set_xlabel("Measure Duration (s)")
axs[3].set_ylabel("Energy Consumption (J)")
axs[3].set_ylim([0, 40])
axs[3].set_xlim([0, 0.9])

plt.suptitle("Energy Consumption vs Measure Duration (Individual Models)", fontsize=14)
plt.tight_layout()
plt.show()
```

### _Trend of varoius perfomance indicators with respect to model_

```python
df = pd.concat([nano_model, small_model, med_model, large_model], axis=0, ignore_index=True)
df.head(5)
```

```python
df1 = df.groupby("current_model", as_index=False)[['input_rate', 'image_quality', 'file_size(kB)', 'avg_conf', 'current_boxes', 'energy_consumption(J)', 'current_cpu', 'measure_duration(s)', 'current_time', 'start_time', 'absolute_time']].mean()
df1.drop(columns=["input_rate", "start_time", "absolute_time"], inplace=True)
df1.sort_values(by=['energy_consumption(J)'], inplace=True)
df1
```
