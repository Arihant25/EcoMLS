#### _epsilon = 0.1_

```python
EcoMLS = pd.read_csv("Results/EcoMLS/log.csv")
EcoMLS["score"] = EcoMLS["energy_consumption(J)"]*(1-EcoMLS["avg_conf"])

EcoMLS["switch"] = (EcoMLS['current_model'] != EcoMLS['current_model'].shift()).astype(int)
EcoMLS.head(5)
```

### NAIVE Approaches_

#### _Approach 1: constant Knowledge_

```python
NAIVE1 = pd.read_csv("Results/NAIVE1/log.csv")
NAIVE1["score"] = NAIVE1["energy_consumption(J)"]*(1-NAIVE1["avg_conf"])

NAIVE1["switch"] = (NAIVE1['current_model'] != NAIVE1['current_model'].shift()).astype(int)
```

#### _NAIVE2: dynamically updating avg conf in Knowledge_

```python
NAIVE2 = pd.read_csv("Results/NAIVE2/log.csv")
NAIVE2["score"] = NAIVE2["energy_consumption(J)"]*(1-NAIVE2["avg_conf"])

NAIVE2["switch"] = (NAIVE2['current_model'] != NAIVE2['current_model'].shift()).astype(int)
```
