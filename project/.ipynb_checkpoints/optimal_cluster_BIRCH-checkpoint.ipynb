```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dfn = pd.read_csv('nano_model_data.csv')
dfs = pd.read_csv('small_model_data.csv')
dfm = pd.read_csv('medium_model_data.csv')
dfl = pd.read_csv('large_model_data.csv')
```

```python
# Store the dataframes
dataframes = [dfn, dfs, dfm, dfl]
model_names = ['nano', 'small', 'medium', 'large']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
def normalize_features(df, feature_columns):
    
    # Extract the features that are to be normalize
    features_to_normalize = df[feature_columns]

    # Initialize the MinMaxScaler
    scaler = MinMaxScaler()

    # Normalize the features
    normalized_features = scaler.fit_transform(features_to_normalize)

    # Create a new DataFrame with the normalized features
    normalized_df = pd.DataFrame(normalized_features, columns=['normalized_energy', 'normalized_avg_conf'])

    # Combine the normalized features with the original DataFrame
    df = pd.concat([df, normalized_df], axis=1)
    
    return df
```

```python
feature_columns = ['energy_consumption(J)', 'avg_conf']

normalized_dataframes = []
for df, model_name in zip(dataframes, model_names):
    normalized_df = normalize_features(df, feature_columns)
    normalized_dataframes.append(normalized_df)
    normalized_df.to_csv(f'{model_name}_model_data_normalized.csv', index=False)
```

```python
from sklearn.cluster import Birch
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
sns.set_style("darkgrid")
sns.set_palette("Set3")
```

```python
def optimal_clusters(df, model_name, selected_features, i):
# df = normalized_dataframes[0]

    # Extract the selected features
    X = df[selected_features].values

    # Preprocess the data if needed (e.g., scale the features)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)


    # Calculate the silhouette scores for different values of k
    silhouette_scores = []
    for k in range(2, 11):
        birch = Birch(n_clusters=k)
        labels = birch.fit_predict(X_scaled)
        silhouette_scores.append(silhouette_score(X_scaled, labels))

    # Plot the silhouette scores
    plt.plot(range(2, 11), silhouette_scores, marker='o')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Silhouette Score')
    plt.title(f'Silhouette Score for Different Numbers of Clusters ({model_name})')
    plt.show()
    
    # Print the silhouette scores for each K value
    for k, s in enumerate(silhouette_scores):
        print(f"K={k+2}, Silhouette score={s:.4f}")
        
    # Determine the optimal K-value using the silhouette scores
    n_clusters[i] = silhouette_scores.index(max(silhouette_scores)) + 2

    print(f"\n\n\033[1mOptimal K-value For BIRCH Clustering (Yolov5 {model_name} Model):\033[0m", n_clusters[i])
```

```python
idx = [0, 1, 2, 3]
n_clusters = [-1 for x in range(4)]
# Select the features
selected_features = ['normalized_energy', 'normalized_avg_conf']

for df, model_name, i in zip(normalized_dataframes, model_names, idx):
    optimal_clusters(df, model_name, selected_features, i)
```

```python
import csv
```

```python
# Create a list of dictionaries with the data
data = [
    {'model': 'Yolov5 Nano', 'n_clusters': n_clusters[0]},
    {'model': 'Yolov5 Small', 'n_clusters': n_clusters[1]},
    {'model': 'Yolov5 Medium', 'n_clusters': n_clusters[2]},
    {'model': 'Yolov5 Large', 'n_clusters': n_clusters[3]}
]

# Open the CSV file for writing
with open('n_clusters.csv', mode='w', newline='') as csv_file:
    # Create a writer object
    writer = csv.DictWriter(csv_file, fieldnames=['model', 'n_clusters'])

    # Write the header row
    writer.writeheader()

    # Write the data rows
    writer.writerows(data)
```
