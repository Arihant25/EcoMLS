```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dfn = pd.read_csv('nano_model_data_normalized.csv')
dfs = pd.read_csv('small_model_data_normalized.csv')
dfm = pd.read_csv('medium_model_data_normalized.csv')
dfl = pd.read_csv('large_model_data_normalized.csv')

df_n_clusters = pd.read_csv("n_clusters.csv")
```

```python
# Store the dataframes and their respective K values
normalized_dataframes = [dfn, dfs, dfm, dfl]

# k_values = df_k_values['k_value'].tolist()
model_names = ['nano', 'small', 'medium', 'large']
```

```python
from sklearn.cluster import Birch
from sklearn.preprocessing import StandardScaler
```

```python
def perform_BIRTCH_clustering(df, n, model_name, selected_features):

    # Extract the selected features
    X = df[selected_features].values

    # Preprocess the data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Initialize the BIRCH clustering algorithm
    birch = Birch(n_clusters=n)

    # Fit the BIRCH model to your data
    birch.fit(X_scaled)

    # Get the cluster labels for each data point and add the cluster labels to DataFrame
    cluster_labels = birch.predict(X_scaled)
    df['BIRCH_Cluster'] = cluster_labels

    # Save the results as CSV files
    df.to_csv(f'{model_name}_BIRCH_Cluster_Results.csv', index=False)

    # Plot the clusters
    plot_clusters(model_name, X_scaled, cluster_labels)

    
def plot_clusters(model_name, X_scaled, cluster_labels):
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=cluster_labels, cmap='viridis')
    plt.title(f'BIRCH Clustering Results ({model_name})')
    plt.xlabel('Energy Consumption (scaled)')
    plt.ylabel('Avg. Confidence (scaled)')
    plt.show()
```

```python
model_names = ['nano', 'small', 'medium', 'large']
n_clusters = df_n_clusters['n_clusters'].tolist()
selected_features = ['normalized_energy', 'normalized_avg_conf']

# Perform clustering and create CSV files for each model
for df, n, model_name in zip(normalized_dataframes, n_clusters, model_names):
    perform_BIRTCH_clustering(df, n, model_name, selected_features)
```

```python
# Read the data from the files
nano_df = pd.read_csv('nano_BIRCH_Cluster_Results.csv')
small_df = pd.read_csv('small_BIRCH_Cluster_Results.csv')
medium_df = pd.read_csv('medium_BIRCH_Cluster_Results.csv')
large_df = pd.read_csv('large_BIRCH_Cluster_Results.csv')
```

```python
# Function to merge dataframes based on the "total_processed" column
def merge_dataframes(df1, df2, suffix):
    return df1.merge(df2, on='total_processed', suffixes=('', suffix))
```

```python
#Nano Model

# List of dataframes and suffixes
dataframes = [nano_df, small_df, medium_df, large_df]
suffixes = ['', '_small', '_medium', '_large']

# Create merged dataframes
merged_nano = nano_df.copy()
for df, suffix in zip(dataframes[1:], suffixes[1:]):
    merged_nano = merge_dataframes(merged_nano, df, suffix)

# Save the merged dataframe as a new CSV file
merged_nano.to_csv('nano_birch_with_other_results.csv', index=False)
```

```python
import os
import scipy.stats as st
import warnings
```

```python
# Import warnings module and filter warnings
warnings.filterwarnings("ignore")

def calculate_confidence_interval(df, columns, cluster_col, alpha=0.1):
    unique_clusters = df[cluster_col].unique()
    confidence_intervals = []
    for cluster in unique_clusters:
        ci_row = [cluster]
        cluster_data = df[df[cluster_col] == cluster][columns].dropna()
        if len(cluster_data) <= 1:
            for col in columns:
                ci_row.append(cluster_data[col].iloc[0])  # Add the same value as lower bound
            for col in columns:
                ci_row.append(cluster_data[col].iloc[0])  # Add the same value as upper bound
        else:
            for col in columns:
                try:
                    ci = st.t.interval(alpha, len(cluster_data[col])-1, loc=np.mean(cluster_data[col]), scale=st.sem(cluster_data[col]))
                    ci_row.append(ci[0])  # Add lower bound
                except:
                    ci_row.append(cluster_data[col].iloc[0])  # Add the same value as lower bound
            for col in columns:
                try:
                    ci = st.t.interval(alpha, len(cluster_data[col])-1, loc=np.mean(cluster_data[col]), scale=st.sem(cluster_data[col]))
                    ci_row.append(ci[1])  # Add upper bound
                except:
                    ci_row.append(cluster_data[col].iloc[0])  # Add the same value as upper bound
        confidence_intervals.append(ci_row)
    ci_columns = [cluster_col] + [f"{col}_CI_Lower" for col in columns] + [f"{col}_CI_Upper" for col in columns]
    return pd.DataFrame(confidence_intervals, columns=ci_columns)
```

```python
input_files = ['nano_birch_with_other_results.csv', 'small_birch_with_other_results.csv',
               'medium_birch_with_other_results.csv', 'large_birch_with_other_results.csv']
main_model = ['nano', 'small', 'medium', 'large']

columns = ['file_size(kB)', 'avg_conf', 'current_boxes', 'energy_consumption(J)',
       'current_cpu', 'measure_duration(s)']
```

```python
for input_file, main_model_name in zip(input_files, main_model):
    
    main_cluster_col = 'BIRCH_Cluster'
    
    output_file = os.path.splitext(input_file)[0] + "_ci.csv"
    
    df = pd.read_csv(input_file)
#     model_names = ['nano', 'small', 'medium', 'large']
    other_models = [m for m in main_model if m != main_model_name]

    main_ci_df = calculate_confidence_interval(df, columns, main_cluster_col)
    ci_dfs = [main_ci_df]

    for other_model in other_models:
        other_columns = [f"{col}_{other_model}" for col in columns]
        other_cluster_col = f"{main_cluster_col}_{other_model}"

        other_ci_df = pd.DataFrame()
        other_ci_df[main_cluster_col] = main_ci_df[main_cluster_col]
        
        for cluster_id in df[main_cluster_col].unique():
            cluster_images = df.loc[df[main_cluster_col] == cluster_id, 'total_processed']
            other_cluster_data = df.loc[df['total_processed'].isin(cluster_images), other_columns]

            for col in other_columns:
                ci = st.t.interval(0.1, len(other_cluster_data[col])-1, loc=np.mean(other_cluster_data[col]), scale=st.sem(other_cluster_data[col]))
                ci_lower_col = f"{col}_CI_Lower"
                ci_upper_col = f"{col}_CI_Upper"
                other_ci_df.loc[other_ci_df[main_cluster_col] == cluster_id, ci_lower_col] = ci[0]
                other_ci_df.loc[other_ci_df[main_cluster_col] == cluster_id, ci_upper_col] = ci[1]

        ci_dfs.append(other_ci_df)

    result = pd.concat(ci_dfs, axis=1)
    result = result.loc[:, ~result.columns.duplicated()]
    result.to_csv(output_file, index=False)
```
